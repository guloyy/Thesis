```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file and skip the first column (contains incrementing number)
csv_file_path = '/Users/defne/Downloads/CAMx_Inorganic_extract_2019 2.csv'
df_csv = pd.read_csv(csv_file_path, usecols=range(1, 8), parse_dates=['Datetime'])


# Drop rows where NO3 or SO4 are missing or a space
df_csv = df_csv[(df_csv['NO3'] != ' ') & (df_csv['SO4'] != ' ')]
df_csv.dropna(subset=['NO3', 'SO4'], inplace=True)

df_csv['Date'] = pd.to_datetime(df_csv['Datetime']).dt.date
df_csv['Date'] = pd.to_datetime(df_csv['Date'])  # Convert to datetime64[ns]


# Group by date and station_id and then take the mean for NO3 and SO4
df_grouped = df_csv.groupby(['Date', 'station_id']).agg({'NO3': 'mean', 'SO4': 'mean'}).reset_index()

# Load Excel file
excel_file_path = '/Users/defne/Downloads/CURRENT DATA REPO/SWITZERLAND_CONSOLIDATED.xlsx'
xl = pd.ExcelFile(excel_file_path)

for sheet in xl.sheet_names:
    df_sheet = xl.parse(sheet, skiprows=[1])  # Skipping the second row with IDs
    df_sheet['Date'] = pd.to_datetime(df_sheet['Date'])  # Convert to datetime64[ns]

    # Extract the ID in the second row of the Station Name column
    id_excel = xl.parse(sheet, nrows=2).iloc[1, 0]

    # Filter records from the grouped CSV DataFrame based on ID
    df_filtered = df_grouped[df_grouped['station_id'] == id_excel]

    # Merge based on Date
    df_merged = pd.merge(df_filtered, df_sheet, how='inner', left_on='Date', right_on='Date')

    # Drop rows where NO3 or SO4 are a space in the merged DataFrame
    df_merged = df_merged[(df_merged['NO3-'] != ' ') & (df_merged['SO42-'] != ' ')]
    df_merged.dropna(subset=['NO3-', 'SO42-'], inplace=True)

    # Time series plot
    plt.figure(figsize=(12, 6))
    plt.plot(df_merged['Date'], df_merged['NO3'], label='Model NO3')
    plt.plot(df_merged['Date'], df_merged['NO3-'], label='Measurement NO3-')
    plt.plot(df_merged['Date'], df_merged['SO4'], label='Model SO4')
    plt.plot(df_merged['Date'], df_merged['SO42-'], label='Measurement SO42-')
    plt.legend()
    plt.title(f'Time Series for ID {id_excel}')
    plt.xlabel('Date')
    plt.ylabel('Concentration')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file and skip the first column (contains incrementing number)
csv_file_path = '/Users/defne/Downloads/CAMx_Inorganic_extract_2019 2.csv'
df_csv = pd.read_csv(csv_file_path, usecols=range(1, 8), parse_dates=['Datetime'])

# Drop rows where NO3 or SO4 are missing or a space
df_csv = df_csv[(df_csv['NO3'] != ' ') & (df_csv['SO4'] != ' ')]
df_csv.dropna(subset=['NO3', 'SO4'], inplace=True)

df_csv['Date'] = pd.to_datetime(df_csv['Datetime']).dt.date
df_csv['Date'] = pd.to_datetime(df_csv['Date'])  # Convert to datetime64[ns]

# Group by date and station_id and then take the mean for NO3 and SO4
df_grouped = df_csv.groupby(['Date', 'station_id']).agg({'NO3': 'mean', 'SO4': 'mean'}).reset_index()

# Load Excel file
excel_file_path = '/Users/defne/Downloads/CURRENT DATA REPO/SWITZERLAND_CONSOLIDATED.xlsx'
xl = pd.ExcelFile(excel_file_path)

for sheet in xl.sheet_names:
    df_sheet = xl.parse(sheet, skiprows=[1])  # Skipping the second row with IDs
    df_sheet['Date'] = pd.to_datetime(df_sheet['Date'])  # Convert to datetime64[ns]

    # Extract the ID in the second row of the Station Name column
    id_excel = xl.parse(sheet, nrows=2).iloc[1, 0]

    # Filter records from the grouped CSV DataFrame based on ID
    df_filtered = df_grouped[df_grouped['station_id'] == id_excel]

    # Merge based on Date
    df_merged = pd.merge(df_filtered, df_sheet, how='inner', left_on='Date', right_on='Date')

    # Drop rows where NO3 or SO4 are a space in the merged DataFrame
    df_merged = df_merged[(df_merged['NO3-'] != ' ') & (df_merged['SO42-'] != ' ')]
    df_merged.dropna(subset=['NO3-', 'SO42-'], inplace=True)

    # Extract PM info from the sheet name
    pm_info = "PM10" if "PM10" in sheet else "PM2.5" if "PM2.5" in sheet else "Unknown"

    # Time series plot
    plt.figure(figsize=(12, 6))
    plt.plot(df_merged['Date'], df_merged['NO3'], label='Model NO3')
    plt.plot(df_merged['Date'], df_merged['NO3-'], label='Measurement NO3-')
    plt.plot(df_merged['Date'], df_merged['SO4'], label='Model SO4')
    plt.plot(df_merged['Date'], df_merged['SO42-'], label='Measurement SO42-')
    plt.legend()
    plt.title(f'Time Series for ID {id_excel} ({pm_info})')  # Include PM info in the title
    plt.xlabel('Date')
    plt.ylabel('Concentration')
    plt.show()
```
